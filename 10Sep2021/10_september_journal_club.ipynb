# Generalization in statistical models

#### This paper exhibits an empirical phenomenon in deep learning that is hard to square with a traditional understanding of statistical learning. But, like, even moreso than usual.

#### Simplified but general classical statistical learning setup
 - A distribution $\mathcal D$ over inputs $x \in \mathcal X$ and labels $y \in \mathcal Y$
 - A *pointwise loss function* $L: \mathcal Y \times \mathcal Y \to \mathbb R$
 - Define the generalization loss of a function $f$ to be $L_D(f) = \mathbb E_{(x, y) \sim D} L(f(x), y)$
 - A hypothesis class $\mathcal H$
 - Goal is to find $f \in \mathcal H$ such that $\mathbb E_{(x, y) \sim \mathcal D} L(f(x), y)$ is minimized
 - A *learning algorithm* $A$ that takes $((x_i, y_i))_{i=1}^m$ in $\mathcal D^m$ for some $m$ and returns $f \in \mathcal H$, hopefully minimizing generalization loss. Typically $A$ will attempt to minimize *training loss* $\sum_{i=1}^m L(f(x_i), y_i)$
 
#### No free lunch theorem
Assume:
- binary classification task
- $m \leq |\mathcal X| / 2$
- $A$ is any learning algorithm

Then there exists a distribution $\mathcal D$ and a function $f : \mathcal X \to \{0, 1\}$ such that $L_D(f) = 0$, but with probability at least $1/7$ over $S \sim \mathcal D^m$, $L_D(A(S)) \geq 1/8$

(Ref.: *Understanding Machine Learning* by S. Ben-David, S. Shalev-Shwartz)

In typical image classification tasks, the number of training images is much smaller than the number of potential inputs, so the NFL theorem in principle applies. Deep neural net models typically have many more parameters than potential inputs, so naively we would expect them to *overfit*. That is, we expect them to perfectly fit the training data but have very poor generalization performance.

In practice, though, deep models exhibit better generalization performance than any other model type on an enormous variety of statistical learning tasks.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(0, 1, 0.025)
Y = X ** 2 + X ** 4 + 0.1 * np.random.normal(size=(40))
plt.scatter(X, Y)
plt.show()
```

```python
deg_3_poly = np.polyfit(X, Y, deg=3)
```

```python
fine_X = np.arange(0, 1, 0.001)
wide_X = np.arange(-1, 2, 0.01)
fine_powers = np.array([fine_X**i for i in range(4)][::-1])
deg_3_fine_preds = np.einsum("i,ij->j", deg_3_poly, fine_powers)
wide_powers = np.array([wide_X**i for i in range(4)][::-1])
deg_3_wide_preds = np.einsum("i,ij->j", deg_3_poly, wide_powers)
```

```python
plt.scatter(X, Y)
plt.plot(fine_X, deg_3_fine_preds)
plt.show()
```

```python
plt.scatter(X, Y)
plt.plot(wide_X, deg_3_wide_preds)
plt.show()
```

```python
deg_40_poly = np.polyfit(X, Y, deg=40)
```

```python
fine_X = np.arange(0, X[-1], 0.001)
fine_powers = np.array([fine_X**i for i in range(41)][::-1])
deg_40_fine_preds = np.einsum("i,ij->j", deg_40_poly, fine_powers)
plt.scatter(X, Y)
plt.plot(fine_X, deg_40_fine_preds)
plt.show()
```

```python
wide_powers = np.array([wide_X**i for i in range(41)][::-1])
deg_40_wide_preds = np.einsum("i,ij->j", deg_40_poly, wide_powers)
plt.scatter(X, Y)
plt.plot(wide_X, deg_40_wide_preds)
plt.show()
```

### Deep models typically do not overfit

- In practice, deep models typically exhibit good generalization performance
- Most natural explanation is that deep learning algorithms are biased to detect physically meaningful features. In vision, convolutional models are popular, and the fact that convolution commutes with translation plausibly produces strong bias for translation-invariant features, for example. Small kernel sizes plausibly bias toward features that depend only on pixel values locally. Convolutional models + whatever exactly SGD does plausibly limits the ability of deep models to fit spurious features, so even though the parameter count is large, the actual hypothesis class of functions that can be learned is much smaller
- This paper all but falsifies this explanation. As far as I know, there does not currently exist a plausible alternative explanation, and in some sense the NFL theorem says that there cannot be any explanation that doesn't involve some inductive bias in the learning algorithm

### Paper results
- It is possible to train deep models to perfect accuracy on training set using random labels
- It is possible to train deep models to perfect accuracy on training set after permuting image pixels, or shuffling image pixels
- These results are not affected by performing standard data augmentations or regularization techniques
- Note, however, that training time is substantially increased

#### What is hard to explain about this?
Consider the following setup:
- Initially, train on both the training data and the validation data, but use true labels for the training data and random labels for the validation data
